```python
import requests
from bs4 import BeautifulSoup
```

```python
# Start with one link: CNN
cnn_url = 'http://www.cnn.com/politics'
cnn = requests.get(cnn_url)
cnn_page = cnn.text
cnn_soup = BeautifulSoup(cnn_page, 'html.parser')
cnn_headline_tags = cnn_soup.findAll('span', 'cd__headline-text')
cnn_headlines = [headline.get_text() for headline in cnn_headline_tags]
cnn_headlines = [headline.encode("utf-8").strip() for headline in cnn_headlines]

print cnn_soup
# with open("head.txt", "wb") as f:
#     f.writelines(["%s" % headline for headline in cnn_headlines])
    
```

```python
# Fox News
# To-do: change unicode chars to what they actually represent

fox_url = 'http://www.foxnews.com/politics.html'
fox = requests.get(fox_url)
fox_page = fox.text
fox_soup = BeautifulSoup(fox_page, 'html.parser')
fox_headline_tags = fox_soup.findAll(itemprop = 'headline')
fox_headlines = [headline.get_text() for headline in fox_headline_tags]
fox_headlines = [headline.encode("utf-8").strip() for headline in fox_headlines]

with open("headlines.txt", "a") as f:
    f.writelines(["%s" % headline for headline in fox_headlines])
```
